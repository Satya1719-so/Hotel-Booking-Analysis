```python
import numpy as np
arr1 =np.array([[10,20],[20,30]])
arr2 =np.array([[20,30],[20,30]])
print(arr1+arr2) # ddition
print(np.add(arr1,arr2)) #another way to addition
print(arr1-arr2)#subtrcat
print(np.subtract(arr1,arr2))
print(arr1*arr2)#multiply
print(np.multiply(arr1,arr2))
print(arr1/arr2)#divide
print(np.divide(arr1,arr2))
```

```python
import numpy as np
arr1 =np.array([[10,20],[20,30]])
arr2 =np.array([[20,30],[40,50]])
print(np.concatenate([arr1,arr2])) # concatenate
print(np.concatenate([arr1,arr2] ,axis=1)) #for vertical axis
print(np.vstack([arr1,arr2])) # vertical
print(np.concatenate([arr1,arr2], axis =0)) #for horizontal
print(np.hstack([arr1,arr2]))
print(np.delete(arr1,1,axis=1))
```
